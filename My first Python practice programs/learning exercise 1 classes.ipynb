```python
# learning exercise classes
import re
import random
dog_db = {}

class Dog:
    age = None  # Damn "None," becomes a (None,) tuple instead of None type...
    name = None
    health = None # Health goes from 0 - 100. 0 Meaning it's dead.
           # Health becomes a random integer on initialization

    def __input_dog_name(self):
        while (True):
            user_input = input("Type in the name of the dog: ")
            if (re.match(r"^[a-zA-Z]+$", user_input)):
                if(dog_db.get(user_input) is None):
                    self.name = user_input
                    return None
                else:
                    print("The name is already in the doggie database!")
                    answer = input("Would you like to change its age?\nN for no, anything else for yes")
                    if (answer != 'N' and answer != 'n'):
                        self.__input_dog_age(dog_db[user_input])
                    return dog_db[user_input]
            else:
                print("The name of the dog must only contain letters.")

    def __input_dog_age(self):
        while (True):
            user_input = input("Type in the age of the dog in dog years: ")
            if (re.match(r"^\d+$", user_input)):
                self.age = int(user_input)
                return
            else:
                print("The age of the dog must only contain numbers.")

    def dog_to_human_age(age):
        assert type(age) == int
        return age // 7

    def human_to_dog_age(age):
        assert type(age) == int
        return age * 7
    
    def __init__(self, dog_name = None, dog_age = None, health = None):
        if (dog_name and dog_db.get(dog_name) is None):
            self.name = dog_name
        elif (dog_name and dog_db.get(dog_name) is not None):
            answer = input("The name is already in the doggie database!\nWould you like to change its age?\nN for no, anything else for yes")
            if (answer != 'N' and answer != 'n'):
                self.__input_dog_age(dog_db[dog_name])
            return dog_db[dog_name]
        else:
            other_dog_obj = self.__input_dog_name()
            if (other_dog_obj is not None): # Cancels rest of initialization
                return other_dog_obj        # Since it's the object is already in the db

        self.age = dog_age or self.age
        if (self.age is None):
            self.__input_dog_age()
        self.health = health or random.randint(0, 100)
        if (self.health <= 20):
            print("The dog sad :(")
# End of class definition

def push_to_db(dog_obj):
    assert isinstance(dog_obj, Dog)
    dog_db[dog_obj.name] = dog_obj

dog_1 = Dog()

push_to_db(dog_1)
for key, value in dog_db.items():
    print(key, value.age)
    
```
