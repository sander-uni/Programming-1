```python
import random
import re
import math

try:
    import matplotlib.pyplot as plt  # Third party module
except ModuleNotFoundError:
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    print("DU TRENGER Å INSTALLERE MATPLOTLIB FOR Å KJØRE DETTE PROGRAMMET.")
    raise

# import numpy
# from numerize import numerize # Third party module

# Skrevet av Sander Østberg

population = 200

MARS_CIV_START_YEAR = 2046
number_of_years = 2000
NUMBER_OF_GRAPHS_TO_MAKE = 1

while(True): # {
    NUMBER_OF_GRAPHS_TO_MAKE = input("Skriv inn hvor mange grafer du vil ha: ")
    if (re.match(r"^\d+$", NUMBER_OF_GRAPHS_TO_MAKE) is None):
        print("Må være et positivt heltall.")
    elif (int(NUMBER_OF_GRAPHS_TO_MAKE) > 5):
        print("Tallet må være 5 eller lavere.")
    else:
        NUMBER_OF_GRAPHS_TO_MAKE = int(NUMBER_OF_GRAPHS_TO_MAKE)
        break
#}

population_data = [None] * 10**5
birth_rate_data = [None] * 10**5
death_rate_data = [None] * 10**5

earth_dr = 7.9  # Global døds rate
earth_br = 17.464  # Global fødsels rate
death_rate = earth_dr * 3
birth_rate = earth_br * 3

for i in range(0, 10**5):# {
    population_death_rate_factor = population / 10**9 # + 1 person ekstra døder per billion populasjonvekst
    random_deathrate_factor = random.randint(-40, 40)

    overpopulation_factor = 1
    if (population > 10**10):
        overpopulation_factor = 2

    if (random_deathrate_factor == 0):
        random_deathrate_factor = 1
        if (random.randint(0, 1) == 1): # 50% sjanse negativ faktor
            random_deathrate_factor * -1
        
    random_deathrate_factor = 1 + (random_deathrate_factor / 100)

    if (random.randint(1, 1000) == 1000):
        population -= random.randint(1, int(min(population * 0.5, 10**7)))

    if (i <= 9):  # Romskip med folk
        population += 200 - random.randint(0, 5)
    population += (population / 1000) * birth_rate  # Fødseler
    population -= \
        (population / 1000) * ( \
            death_rate * \
            population_death_rate_factor * \
            random_deathrate_factor * \
            overpopulation_factor \
        )  # Døder
    
    death_rate = death_rate >= (earth_dr / 2) and death_rate - min(0.01, 0.001 * i) or death_rate
    
    if (i > 9):
        birth_rate = birth_rate >= earth_br and birth_rate - min(0.03, 0.003 * i) or birth_rate
    
    population_data[i] = population
    death_rate_data[i] = death_rate
    birth_rate_data[i] = birth_rate
# }

# Generate year datapoints
year_axis = [None] * 10**5
for i in range(MARS_CIV_START_YEAR, MARS_CIV_START_YEAR + 10**5):
    year_axis[i - MARS_CIV_START_YEAR] = i

for _ in range(0, NUMBER_OF_GRAPHS_TO_MAKE):
    while(True): # {
        number_of_years = input("Skriv inn hvor mange år du skal simulere: ")
        if (re.match(r"^\d+$", number_of_years) is None):
            print("Må være et positivt heltall.")
        elif (int(number_of_years) > 10**5):
            print("Tallet må være 100 tusen eller lavere.")
        else:
            number_of_years = int(number_of_years)
            break
    #}

    local_pop_data = population_data[0:number_of_years:1]
    local_year_data = year_axis[0:number_of_years:1]

    figure, axes = plt.subplots()

    axes.set_title("Populasjonsvekst på Mars")
    axes.set_xlabel("År")
    axes.set_ylabel("Populasjon")

    axes.plot(local_year_data, local_pop_data) # Populasjons graf

#    y_tick_labels = [200, 200]
#    y_tick_labels_original = axes.get_yticklabels()
#    largest_data_point = max(local_pop_data)
#    for i in range(3, axes.get_yticks().size):
#        y_tick_labels.append(f"{numerize.numerize(largest_data_point / (axes.get_yticks().size - i))}")
#
#    axes.set_yticklabels(labels = y_tick_labels)

    x_ticks = axes.get_xticks()
    x_ticks[0] = "2046"
    x_ticks[1] = "2046"
    axes.set_xticks(ticks = x_ticks)

    plt.show()
```
